Vamos começar analisando o modelo de progrmação clássico. Para isso, vamos criar uma função que receberá um número (x) e retornará o número recebido multiplicado por 2 e subtraído de 1 
[(x*2)-1]

```python
def multiplicaPorDoisSubtraiUm(x):
  y = (x*2)-1
  return y
```

```python
multiplicaPorDoisSubtraiUm(10)
```

Acima, utilizamos o modelo tradicional de programação, onde desenvolvemos o script e passamos os dados para a máquina. Agora, vamos fazer diferente, vamos utilizar o modelo clássico de Inteligência Artificial. Vamos passar para a máquina dados com os resultados correspondentes para ver se a máquina cria bem os procedimentos.

```python
import tensorflow as tf
```

 Uma vez que o TensorFlow foi importado, vamos criar uma rede neural simples. Esta rede contemplará UMA camada e UM neurônio.

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
```

Uma vez que o neurônio foi criado, vamos passar os parâmetros de configuração.

```python
model.compile(optimizer="sgd", loss="mean_squared_error")
```

Criamos uma rede neural que possui UMA camada com UM neurônio e com UMA entrada. 
Compilamos a rede configurando a medida de perda por meio do cálculo do mínimo erro ao quadrado (mean_squared_erros). Também configuramos a função para reduzir o erro até que o modelo possa atingit o melhor resultado, essa função é o Estocástico Gradiente Descendente.

Agora, vamos gerar os dados que serão utilizados no treinamento.

```python
import numpy as np
X = np.array([-1.0, 0.0, 3.0, 4.0, 5.0])
y = np.array([-3.0, -1.0, 5.0, 7.0, 9.0]) #x*2-1
```

Uma vez que temos os valores de entrada com as respectivas saídas, vamos treinar a rede neural para que ela seja capaz de advinhar a fórmula que levará ao resultado.

```python
model.fit(X,y, epochs=500)
```

Uma vez que o modelo foi treinado, vamos ver se ele deduziu bem a lógica por tráz dos dados.

```python
model.predict([100])
```
